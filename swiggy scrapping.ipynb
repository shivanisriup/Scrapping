Scrapping of swiggy website using beautifulsoup in python 

```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.swiggy.com/varanasi"
r = requests.get(URL) 
  
soup = BeautifulSoup(r.content, 'html5lib') 
print(soup.prettify())
```

```python
quotes=[]  # a list to store quotes 
  
table = soup.find('div', attrs = {'id':'restaurants_container'}) 
#print(table)
  
for row in table.findAll('div', attrs = {'class':'MZy1T'}):
    #print(row)
    quote = {} 
    quote['url'] = row.a['href'] 
    quote['Title'] = row.img['alt'] 
    quote['src'] = row.find('div', attrs = {'class':'_1gURR'}).text
    quote['cost']=row.find('div', attrs = {'class':'nVWSi'}).text
    quotes.append(quote) 
```
